```python
import py_trees
import sys
import time

import py_trees.console as console

def create_root():
    root = py_trees.composites.Sequence("Sequence")
    for action in ["Action 1", "Action 2", "Action 3"]:
        success_after_two = py_trees.behaviours.Count(name=action,
                                                      fail_until=0,
                                                      running_until=1,
                                                      success_until=2)
        root.add_child(success_after_two)
    return root
```

```python
def create_root1():
    root = py_trees.composites.Sequence("Sequence")
    action = "Action 1"
    node = py_trees.behaviours.Count(name=action,
                                        fail_until=0,
                                        running_until=1,
                                        success_until=2
                                    )
    root.add_children([node])
    
    return root
```

```python
root = create_root1()

root.setup_with_descendants()
for i in range(1, 5):
    try:
        print("\n--------- Tick {0} ---------\n".format(i))
        root.tick_once()
        print("\n")
        print(py_trees.display.unicode_tree(root=root, show_status=True))
        time.sleep(0.5)
    except KeyboardInterrupt:
        break
print("\n")
```
